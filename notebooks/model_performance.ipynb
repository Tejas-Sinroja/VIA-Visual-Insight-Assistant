```python
import pickle
import matplotlib.pyplot as plt

model_histories = {}
for i in range(1, 5):
    with open(f'D:\\VIA\\Project\\app\\models\\model\\cnnv{i}.pkl', 'rb') as file:
        model_histories[f'cnnv{i}'] = pickle.load(file)

for model_name, history in model_histories.items():
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history['accuracy'], label='Training Accuracy')
    plt.plot(history['val_accuracy'], label='Validation Accuracy')
    plt.title(f'{model_name} Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history['loss'], label='Training Loss')
    plt.plot(history['val_loss'], label='Validation Loss')
    plt.title(f'{model_name} Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
for model_name, history in model_histories.items():
    plt.plot(history['val_accuracy'], label=f'{model_name} Validation Accuracy')
plt.title('Validation Accuracy Comparison')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
for model_name, history in model_histories.items():
    plt.plot(history['val_loss'], label=f'{model_name} Validation Loss')
plt.title('Validation Loss Comparison')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pickle
import matplotlib.pyplot as plt

with open('D:\\VIA\\Project\\app\\models\\model\\improved_cnnv2.pkl', 'rb') as file:
    history_cnnv2 = pickle.load(file)

with open('D:\\VIA\\Project\\app\\models\\model\\improved_cnnv21.pkl', 'rb') as file:
    history_cnnv21 = pickle.load(file)

epochs_cnnv2 = range(1, len(history_cnnv2['accuracy']) + 1)
epochs_cnnv21 = range(1, len(history_cnnv21['accuracy']) + 1)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs_cnnv2, history_cnnv2['accuracy'], label='Train Accuracy', marker='o')
plt.plot(epochs_cnnv2, history_cnnv2['val_accuracy'], label='Validation Accuracy', marker='x')
plt.title('CNNv2: Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epochs_cnnv2, history_cnnv2['loss'], label='Train Loss', marker='o')
plt.plot(epochs_cnnv2, history_cnnv2['val_loss'], label='Validation Loss', marker='x')
plt.title('CNNv2: Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs_cnnv21, history_cnnv21['accuracy'], label='Train Accuracy', marker='o')
plt.plot(epochs_cnnv21, history_cnnv21['val_accuracy'], label='Validation Accuracy', marker='x')
plt.title('CNNv21: Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epochs_cnnv21, history_cnnv21['loss'], label='Train Loss', marker='o')
plt.plot(epochs_cnnv21, history_cnnv21['val_loss'], label='Validation Loss', marker='x')
plt.title('CNNv21: Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(epochs_cnnv2, history_cnnv2['accuracy'], label='improved_CNNv2 Train Acc', marker='o')
plt.plot(epochs_cnnv2, history_cnnv2['val_accuracy'], label='improved_CNNv2 Val Acc', marker='x')
plt.plot(epochs_cnnv21, history_cnnv21['accuracy'], label='improved_CNNv21 Train Acc', marker='o', linestyle='dashed')
plt.plot(epochs_cnnv21, history_cnnv21['val_accuracy'], label='improved_CNNv21 Val Acc', marker='x', linestyle='dashed')
plt.title('Training and Validation Accuracy Comparison')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(epochs_cnnv2, history_cnnv2['loss'], label='improved_CNNv2 Train Loss', marker='o')
plt.plot(epochs_cnnv2, history_cnnv2['val_loss'], label='improved_CNNv2 Val Loss', marker='x')
plt.plot(epochs_cnnv21, history_cnnv21['loss'], label='improved_CNNv21 Train Loss', marker='o', linestyle='dashed')
plt.plot(epochs_cnnv21, history_cnnv21['val_loss'], label='improved_CNNv21 Val Loss', marker='x', linestyle='dashed')
plt.title('Training and Validation Loss Comparison')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
